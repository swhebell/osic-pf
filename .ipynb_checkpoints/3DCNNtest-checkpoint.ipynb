```python
%load_ext tensorboard

import datetime

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras import datasets, layers, models

from sklearn.model_selection import train_test_split
```

```python
# GPU options to limit OOM erors
physical_devices = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
# Clear old logs for tensorboard
!rm -rf ./logs/ 
```

```python
measures = pd.read_csv('osic-pulmonary-fibrosis-progression/train.csv')
```

```python
# make a list of unique patients
patients = measures['Patient'].unique()
```

```python
# Compute the mean FVC for each patient and store it in a new dataframe
meanfvc = pd.DataFrame(index=patients, columns=['mean_FVC'])
for i in patients:
    meanfvc.loc[i, 'mean_FVC'] = np.round(measures.FVC[measures['Patient'] == i].mean(), 0).astype('int16')
```

```python
# Load the CT DF
pfdf = pd.read_pickle('ctdf.pkl')
```

```python
# Add a the mean FVC to each patient we have a CT for
for i in pfdf.index:
    pfdf.loc[i, 'mean_FVC'] = meanfvc.loc[i, 'mean_FVC']
```

```python
# Check to see if we managed to save and then load the CTs correctly
fig = plt.figure(figsize=[15,15])
for i in range(20):
    y = fig.add_subplot(4,5,i+1)
    y.imshow(pfdf.iloc[56,0][i], cmap='gray')
    
```

```python
# Features and labels
X = np.array(list(pfdf['ct']))
y = np.array(pfdf['mean_FVC'])
```

```python
X = X.reshape(171,20,200,200,1)
```

```python
# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
X_train.shape
```

```python
model = models.Sequential()

model.add(layers.Conv3D(32, (3, 3, 3), activation='relu', padding='same', data_format='channels_last',
                        input_shape=(20,200,200,1))) # set input shape to dimensions chosen for CT imaging
model.add(layers.MaxPooling3D((2, 2, 2)))
model.add(layers.Conv3D(128, (3, 3, 3), activation='relu', padding='same'))
model.add(layers.MaxPooling3D((2, 2, 2)))
model.add(layers.Conv3D(256, (3, 3, 3), activation='relu', padding='same'))
model.add(layers.MaxPooling3D((2, 2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu',
                       kernel_regularizer=tf.keras.regularizers.l1(0.01)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='linear'))


#kernel_regularizer=tf.keras.regularizers.l1(0.001)
```

```python
model.summary()
```

```python
# Compile the model, set optimiser, loss function and metrics
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001),
             loss='mae',
             metrics=['mae', 'mse'])
```

```python
# Specify callbacks for tensorboard
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
history = model.fit(X_train, y_train, epochs=40,
                    validation_data=(X_test, y_test), batch_size = 2,
                    callbacks=[tensorboard_callback])
```

```python
# Plot the training and validation loss

hist = pd.DataFrame(history.history)

plt.figure()
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")

plt.plot(hist.index[1:], hist['mae'][1:], label="Train Loss")
plt.plot(hist.index[1:], hist['val_mae'][1:], label="Val Loss")
plt.legend()
plt.show() 
```

```python
# Look for bias

predictions = model.predict(X_test, batch_size=1).flatten()
a = plt.axes(aspect='equal')
plt.scatter(predictions, y_test, edgecolors=(0, 0, 0))
plt.xlabel('True Values')
plt.ylabel('Predictions')
lims = [1000, 4000]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```
